```python
def thresh_cut(results, t):
    return results[results.threshold == t]

def preproc_f(dataset):
    thres_1 = dataset
    index_scaled = preprocessing.scale(thres_1['index'])
    thresh_scaled = preprocessing.scale(thres_1['thresh'])
    npix_scaled = preprocessing.scale(thres_1['npix'])
    tnpix_scaled = preprocessing.scale(thres_1['tnpix'])
    xmin_scaled = preprocessing.scale(thres_1['xmin'])
    xmax_scaled = preprocessing.scale(thres_1['xmax'])
    ymin_scaled = preprocessing.scale(thres_1['ymin'])
    ymax_scaled = preprocessing.scale(thres_1['ymax'])
    x_scaled = preprocessing.scale(thres_1['x'])
    y_scaled = preprocessing.scale(thres_1['y'])
    x2_scaled = preprocessing.scale(thres_1['x2'])
    y2_scaled = preprocessing.scale(thres_1['y2'])
    xy_scaled = preprocessing.scale(thres_1['xy'])
    a_scaled = preprocessing.scale(thres_1['a'])
    b_scaled = preprocessing.scale(thres_1['b'])
    theta_scaled = preprocessing.scale(thres_1['theta'])
    cxx_scaled = preprocessing.scale(thres_1['cxx'])
    cyy_scaled = preprocessing.scale(thres_1['cyy'])
    cxy_scaled = preprocessing.scale(thres_1['cxy'])
    cflux_scaled = preprocessing.scale(thres_1['cflux'])
    flux_scaled = preprocessing.scale(thres_1['flux'])
    cpeak_scaled = preprocessing.scale(thres_1['cpeak'])
    peak_scaled = preprocessing.scale(thres_1['peak'])
    xcpeak_scaled = preprocessing.scale(thres_1['xcpeak'])
    ycpeak_scaled = preprocessing.scale(thres_1['ycpeak'])
    xpeak_scaled = preprocessing.scale(thres_1['xpeak'])
    ypeak_scaled = preprocessing.scale(thres_1['ypeak'])
    flag_scaled = preprocessing.scale(thres_1['flag'])
    objID_scaled = preprocessing.scale(thres_1['objID'])
    cstarid_scaled = preprocessing.scale(thres_1['cstarid'])
    threshold_scaled = preprocessing.scale(thres_1['threshold'])
    cstarmag_scaled = preprocessing.scale(thres_1['cstarmag'])
    was_a_hit_scaled = preprocessing.scale(thres_1['was_a_hit'])
    
    X = np.array([npix_scaled, tnpix_scaled, xmin_scaled, xmax_scaled,ymin_scaled,ymax_scaled,x_scaled,
              y_scaled,x2_scaled,y2_scaled,xy_scaled,a_scaled,b_scaled,theta_scaled,cxx_scaled,cyy_scaled,
              cxy_scaled,cflux_scaled,flux_scaled,cpeak_scaled,peak_scaled,xcpeak_scaled,ycpeak_scaled,
              xpeak_scaled,ypeak_scaled,flag_scaled]).T

    Y = np.asarray((thres_1['was_a_hit']==True).astype(int) ) #rbdata[0]['real_bogus']
    
    return (X, Y)
```

```python
from sklearn import metrics, cross_validation

class RunMLTest(object):
    """Class to run a complete ML test using a set of parameters.
    Different classifiers offer different parameters, and so this 
    class runs a test assuming you have settled these values correctly.
    """
    def __init__(self, data_X, data_Y, classifier_instance, nfolds):
        self.X = data_X
        self.Y = data_Y
        self.clf = classifier_instance
        self.nfolds = nfolds
        
        
    def _skf(self):
        self.skf = cross_validation.StratifiedKFold(self.Y, self.nfolds)
    
    
    def run(self):
        self._skf()
        
        self.precision = cross_validation.cross_val_score(self.clf, self.X, self.Y, cv=self.skf, scoring='precision')
        self.recall = cross_validation.cross_val_score(self.clf, self.X, self.Y, cv=self.skf, scoring='recall')
        self.f1 = cross_validation.cross_val_score(self.clf, self.X, self.Y, cv=self.skf, scoring='f1')
        
        self.mean_scores = {'P': self.precision.mean(),
                            'R': self.recall.mean(),
                            'F1': self.f1.mean()}
        
    
```

```python
def classifier_test(thresholds, clf, nfolds):
    scores = []
    for t in thresholds:
        X, Y = preproc_f(thresh_cut(results, t))
        
        run = RunMLTest(X, Y, clf, nfolds)
        run.run()
        scores.append(run.mean_scores)
    return scores
```
